## **1. Pearson Correlation**

The Pearson correlation is the most used correlation measure. The formula for Pearson Correlation is:

$$ \rho_{X,Y} = \frac{cov(X,Y)}{\sigma_{_{X}} \sigma_{_{Y}}} $$

Where $cov(X,Y)$ is the covariance of $X$ and $Y$, $\sigma_{_{X}}$ is the standard deviation of $X$ and $\sigma_{_{Y}}$  is the standard deviation of $Y$

With some math, we can also write the Pearson correlation as follows:

$$ \rho_{X,Y} = \frac{E(XY)-E(X)E(Y)}{ \sqrt{E(X^{2})-(E(X))^{2}} \sqrt{E(Y^{2})-(E(Y))^{2}}} $$

Pearson correlation is a measure of the linear relationship that exists between two random variables. For example, we can calculate the Pearson correlation of hours of studying and test grades for 50 students to see if there is a linear correlation. In order to calculate Pearson correlation, you must have paired values from each observation. The value of Pearson Correlation is between $-1$ and $1$. In general, if the absolute value of the Pearson correlation for two random variables is over 0.7, we would say the two variables are highly correlated. If the absolute value of the Pearson correlation for two random variables is between 0.5 and 0.7, we would say the two variables are moderately correlated. If the Pearson correlation is between 0.3 and 0.5, we would say the two variables are low correlated. However, we would say the two variables are not correlated when the Pearson correlation is below 0.3. Because the Pearson Correlation only measures the "linear relationship" of two variables, a low Pearson Correlation only says that the two variables are not linearly correlated, but they may be correlated in a non-linear way. The Pearson correlation also assumes that the two variables come from normal distribution and are continuous variables. 

One problem with the Pearson correlation is that it can change significantly when outliers are present. A measure whose value does not change significantly with the existence of outliers is called a robust measure. For example, the median is a **robust measure**. In a dataset $\{2, 2, 2, 3, 5, 6, 6, 7, 8, 9, 10, 5000 \}$, 5,000 is the outlier. The median is six if you keep the outlier. The median is still six if you drop the outlier. The story is different for the mean. The mean is 421.67 with the outlier. The mean is 5.45 without the outlier. The mean decreases significantly without the outlier while the median does not change without the outlier.

From figure 1, you can see that the Pearson Correlation changes significantly if there is an outlier in the dataset. The Pearson Correlation is not resistant to outliers. Due to this issue, we need to find more robust measures to calculate the correlation of two random variables when there are outliers in the dataset. In the following sections, we will introduce the Spearman correlation and Kendall correlation to handle the issue of outliers in calculating correlation.


## **2. Spearman Correlation**

The Spearman correlation $\rho$ is also a measure of the relationship between two random variables. However, instead of using the data values to calculate correlation, the Spearman correlation is a non-parametric method that uses ranks to measure correlation. Lets use the following example to demonstrate how the Spearman correlation is calculated.

In figure 2, we have paired data from 15 students about how many days they study a subject and the test grade they received. To calculate the Pearson correlation, we use the "Hours of Study" column and "Test Grade" column. However, to calculate the Spearman correlation, we first need to rank each random variable. For the "Hours of Study" random variable, we rank the observations by giving 1 to the highest "Hours of Study" and 2 to the next highest "Hours of Study." Then, we do the same for the "Test Grade" random variable. From figure 2, we can see the two columns on the right are ranked "Hours of Study" and "Test Grade." We then apply these two columns to the Pearson correlation formula to get the Spearman correlation. Hence, we can say that the Spearman correlation is the ranked version of the Pearson correlation. Now we know that the Spearman correlation is the ranked version of the Pearson correlation and that the Spearman correlation is less sensitive to outliers. Lets look at other properties of the Spearman correlation.

> **a. The value of the Spearman correlation is between -1 and 1.** <br>
>  This conclusion is the result of using the Pearson correlation formula on ranked values of the data.
>
> **b. The Spearman correlation is a ranked data correlation.** <br>
>  As demonstrated above, if the data is ranked data, you should use the Spearman correlation.
>
> **c. The Spearman correlation is resistant to outliers.** <br>
>  We mentioned this point above.
>
> **d. The Spearman correlation is a non-parametric method.** <br>
>  Since Spearman correlation does not require the data to be normal, it is considered a non-parametric method. 
>
> **e. The Spearman correlation is a measure of the strength of the monotonic relationship between two random variables.** <br>

Because Spearman correlation measures the monotonic relationship of two variables, there is no assumption of a linear relationship of two variables. ## **3. Kendall Correlation**

Kendall Correlation or Kendalls tau Correlation $\tau$ is another non-parametric rank-based correlation. To calculate Kendalls tau Correlation, we first need to determine concordant numbers and discordant numbers. Concordant observations mean the values of two variables are moving in the same direction. Discordant observations mean the values of two variables are moving in the opposite direction. It is best to use an example to illustrate how to get the numbers.


**Figure 7: Kendalls tau Correlation Data Example**

| Student | Hours of Study | Test Grade | Ranked Hours | Ranked Grade | Concordant | Discordant |
| :---: | :---: | :---: | :---: | :---: | :---: | :---: |
|   I   |   15  |   84  |   1   |   3   |   9   |   2   |
|   A   |   14  |   89  |   2   |   2   |   9   |   1   |
|   D   |   13  |   91  |   3   |   1   |   9   |   0   |
|   L   |   13  |   35  |   4   |  12   |   0   |   8   |
|   J   |   12  |   75  |   5   |   4   |   7   |   0   |
|   F   |   11  |   71  |   6   |   7   |   4   |   2   |
|   K   |   10  |   74  |   7   |   5   |   5   |   0   |
|   G   |    9  |   70  |   8   |   8   |   3   |   1   |
|   B   |    8  |   72  |   9   |   6   |   3   |   0   |
|   E   |    7  |   59  |  10   |   9   |   2   |   0   |
|   H   |    6  |   42  |  11   |  11   |   0   |   1   |
|   C   |    4  |   47  |  12   |  10   |       |       |
|       |       |       |       |       |----------------- <br> **51** | ----------------- <br> **15** |



In figure 7, we show the same example of a students study hours and their test grades. We first ordered the table with the student who has the highest "Hours of Study" on top and the least on the bottom. Then, we added a few more columns. For "Ranked Hours," we use 1 for the student who has the highest "Hours of Study" and 2 for the second highest "Hours of Study." We repeat this process with the "Ranked Grade" column. Because the table is ordered by "Hours of Study," you can see "Ranked Hours" has a starting value of 1 and an ending value of 12.

Now we'll examine how to get the value under the "Concordant" column. First, we start with the first row for Student I. Remember the table is already ordered by the first variable "Hours of Study." Lets focus on the "Ranked Grade" column. For student I, the "Ranked Grade" is 3. We need to count how many numbers are larger than 3 below the row where 3 sits. In this case, there are 9 numbers {12, 4, 7, 5, 8, 6, 9, 11, 10} that are larger than 3. We put 9 in the "Concordant" cell in the row where 3 sits. We then move to the next row for student A. The number for "Ranked Grade" is 2. We need to calculate how many numbers are greater than 2 below this row. There are 9 numbers {12, 4, 7, 5, 8, 6, 9, 11, 10}. We put 9 in the "Concordant" cell to the row where 2 sits. We keep doing it until we reach the row of student C. Since there are no numbers under student C, the "Concordant" cell is left empty.

Now lets move to the "Discordant" column. We also start from the first row for student I. We still focus on the "Ranked Grade" column. We need to count how many numbers are smaller than 3. In this case, there are 2 numbers {1, 2}. We put 2 in the "Discordant" cell to the row where 3 sits. Then, we keep doing the same for the following row until the last row. Then, sum up all the numbers in both the "Concordant" column and "Discordant" column.

Kendalls tau Correlation is defined as follows:

$$ \tau = \frac{\text{Number of Concordant} - \text{Number of Discordant}}{\binom{n}{2}} $$

Where $n$ is number of observations

Apply our example to the formula:

$$ \tau = \frac{51-15}{\binom{12}{2}} = \frac{36}{66} = 0.54545 $$

The above Kendalls tau is also called Kendalls tau-a. There are alternative definitions, like Kendalls tau-b, which adjusts for tied ranks. 

Here are some other properties for Kendalls tau:

> 1. The value of Kendalls tau is between $-1$ and $1$. 
> 2. Kendalls tau is also a non-parametric statistic since it does not require the variables to be normally distributed. 
> 3. Kendalls tau does not require the two variables to have a linear relationship.
> 4. Kendalls tau is also resistant to outliers.

Both Kendalls tau and the Spearman correlation are rank correlations. In general, the Spearman correlation tends to be larger than Kendalls tau in large sample sizes. The Spearman correlation is also more sensitive to data error or discrepancy. Hence, Kendalls tau is generally preferred over the Spearman correlation because Kendalls tau is more robust and more efficient. Kendalls tau is also easier to explain using the difference between concordance and discordance. Given both methods are in most statistical applications, you can run both and compare the results. 

## **4. Concordant Correlation**

In this section, we are going to introduce a correlation metric to evaluate agreement. The correlation metrics we have discussed so far all focus on correlation. But what is the difference between correlation and agreement? We will talk about the difference in this section and then introduce concordant correlation when looking for agreement as the research goal.


### **4.1 Correlation vs. Agreement**

To determine correlation is to measure the strength of association between two variables. These two variables do not need to be the same subjects or on the same scale. For example, we would like to know the association between a students study hours and their test grades. These are two different variables with different measurement units. 

To determine agreement is also to measure the strength of association between two variables. However, it focuses on degree of concordance. For example, two doctors assessing a group of patients might like to see if their assessments agree or not. Another example is two teachers critiques on students' writing. The two variables should be the same construct and of course the same measurement. Sometimes, agreement is also related to whether the tests are reproducible (reliable). Hence, agreement is also called reproducibility. For agreement analysis, we not only need to know how two variables are correlated but we also need to know the bias from different opinions or rating systems. For agreement analysis or concordance analysis, we do not need the variables to follow a linear relationship. What we would like to have is for the values of two variables for an observation to be close to each other, which means the opinion or rating is similar. This is why when we analyze the correlation of this kind of assessment/opinion variable, we will need a new correlation. Concordance correlation is useful for this reason.


### **4.2 Basics of Concordance Correlation**

Lawrence Lin proposes how to use the concordance correlation coefficient. Lets see how it works. If we have two random variables $x$ and $y$, here is the formula for concordance correlation:

$$ P_{ccc} = \frac{2 \sigma_{xy}}{\sigma_{x}^{2} + \sigma_{y}^{2} + (\mu_{x} - \mu_{y})^{2}} $$

Where $\sigma_{xy}$ is the covariance, $\sigma_{x}$ and $\sigma_{y}$ are variances for $x$ and $y$, and $\mu_{x}$ and $\mu_{y}$ are expected values (means) for $x$ and $y$.

We can decompose the above formula as follows:

$$ P_{ccc} = \rho \times C_{b} $$

Where 

$$ C_{b} = \frac{2}{\frac{\sigma_{1}}{\sigma_{2}} + \frac{\sigma_{2}}{\sigma_{1}} + \frac{(\mu_{1}-\mu_{2})^{2}}{\sigma_{1}\sigma_{2}}} $$ 

is bias correction factor and $\rho$ is Pearson Correlation.

Here are some properties of Concordance Correlation:

> 1. If $\rho = 1$, $\mu_{x} = \mu_{y}$ and $\sigma_{x} = \sigma_{y}$, then $ P_{ccc} = 1$ and $x$ and $y$ are perfectly positively in agreement.
> 2. Concordance correlation is between $-1$ and $1$.
> 3. $C_{b}$ bias correction factor is between $0$ and $1$.
> 4. The smaller the $C_{b}$ bias correction factor, the larger the bias is. From the bias correction factor formula, we can see that as the two variables get a higher degree of disagreement, the denominator of the bias correction factor will get bigger, but the numerator will stay the same. Hence, the bias correction factor will get smaller as the disagreement gets worse.
> 5. A low agreement between two variables can stem from low correlation (low $\rho$) and/or high bias (low bias correction factor).


## **5. Asset Returns and Correlations**

Now lets apply the correlation methods we discussed in this lesson to the U.S. 10-Year Treasury Bond Yield daily return and Dow Jones Industrial Average daily return.


**Figure 8: Scatter Plot for U.S. 10-Year Treasury Bond Yield Daily Return and Dow Jones Industrial Average Daily Return**


```python
# Scatter plot for DWJ and 10Y_TBY
scatt = data_set.plot.scatter(x="10Y_TBY", y="DWJ")
scatt.set(xlabel="10Y_TBY", ylabel="DWJ")
plt.show()

# Pearson, Spearman and Kendal tau correlation coefficients
corr_pearson = pearsonr(data_set["10Y_TBY"], data_set["DWJ"])
corr_spearman = spearmanr(data_set["10Y_TBY"], data_set["DWJ"])
corr_kendall = kendalltau(data_set["10Y_TBY"], data_set["DWJ"])
print(
    "Correlation Pearson     : ",
    corr_pearson[0].round(4),
    "\nCorrelation Spearman    : ",
    corr_spearman.correlation.round(4),
    "\nCorrelation kendalltau  : ",
    corr_kendall.correlation.round(4),
)
```

From figure 8 and the above correlation results, we can see that there are outliers in the dataset. Hence, the Pearson Correlation is probably not the best choice. As expected, the Spearman correlation has a larger value than Kendalls tau. However, since the correlation is just around 0.3 depending on whether you're using the Spearman correlation or Kendalls tau, the two variables have low correlation.
