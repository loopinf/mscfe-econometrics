## **1. Supervised Learning**

**Supervised Learning** means that a model is constructed to predict data with label. What does it mean in plain English? For example, if there 10 cats, 10 dogs and 10 chickens in the assigned dataset,  for these 30 animals, we also have a description of their feature variables, like the shape of their noses, the shape of their mouth, fur type, body type etc… We would like to build a model using feature variables as input variables to predict if an animal is a cat, a dog or a chicken. Once the model is created, we are able to feed the features of a certain animal to this model and the model tells us what this animal is. This simple modeling concept is the same concept of a more complicated facial recognition model used in Apple’s iPhone. 

Labels data can be numeric or categorical. Our last animal example is a categorical example. One example for a supervised learning model with numeric labels of data is used to predict someone’s income level. Assume we have 100 people with their income data, education level and other variables. In this case, to each person is attached a label that contains his or her  income level. We would like to build a model based on certain information we have to predict income level. Does this model sound familiar? This is a regression model. Regression models that we have considered so far are part of the supervised learning in the field machine learning. 

With the above examples, we can see that there are two types of supervised learning models: **Classification and Regression**. 

> **Classification models:** models for data with categorical label or with categorical output variable. For example, types of animals, positive or negative for a specific form of cancer, identify stocks as value or growth stocks, etc… Popular classification models include discriminant analysis, logistic regression, K nearest neighbors and naïve Bayes classifiers, decision tree and support vector machine.

> **Regression models:** models for data with numerical label or with numerical output variable. For example, income level, stock return, GDP growth etc… Linear regression is the most popular one in this category.


## **2. Unsupervised Learning**

Sometimes, for a data project, we may not have a target variable or an output variable that we want to predict. Instead, we would like to analyze the data and see if we can group the data according to a particular pattern, similarity or features etc... We can also say that with unlabeled data, we would like to use the information we have from the data to assign it to certain groups. Models that aim at grouping unlabeled data are defined as **Unsupervised Learning**.  These models need to find the hidden pattern within the data in order to separate the data into group elements. 

For instance, we can collect trading information from traders in the stock market. For example, we could collect information like types of stock traded, position holding duration, strategies used etc… With all this information, we can use an unsupervised learning model to identify traders with similar behaviors and group them together. After the grouping is done, we can investigate traders’ behavior variables within the same group and decide what are the common traits inside that particular group. After investigating behavior variables, we may come up with  certain particular groups like buy-and-hold traders, high frequency traders or technical traders etc…

Popular unsupervised learning models include hierarchical clustering, k-means clustering, and principal component analysis. 

Supervised learning models and unsupervised learning models are two major modeling tools in machine learning. Under each modeling group, one can identify several modeling methods. In the following sections, we will introduce discriminant analysis from the supervised learning modeling group and clustering analysis from the unsupervised learning modeling group.


## **3. Discriminant Analysis**

**Discriminant analysis (DA)** is a type of model used for supervised learning modeling. DA is used to predict the category of the subject given certain input data about the subject. One example we talked about in last section is to predict if an animal is a cat, a dog or a chicken given this animal’s feature information. Hence, we also call DA a **Classification** model. A DA model is only used for analysis with categorical output. 

A very simple example is beneficial to explain how DA works. Assume we have 70 active traders and 70 passive traders. We also have the data of their position holding duration in days and average number of trades they do per year. Our goal is to see if a trader’s position holding duration and number of trades per year can tell us what type of trader  he or she belongs to.

In Figure 1, we can see that active traders usually have higher number of trades but short during for position holding and passive traders do the reverse. However, how can we use these two input variables to better predict the type of a trader? We can use DA in this case. Specifically, we will use **Linear Discriminant Analysis (LDA)**. 


### **3.1. Linear Discriminant Analysis (LDA): Theory**

**Linear Discriminant Analysis (LDA)** provides a method to assign or classify a subject whose type is unknown to a type based on particular input data for this subject. In our example above, if we have a trader that we don’t know whether he/she is an active or a passive trader, but we have the data regarding this trader’s holding duration and number of trades, we can use LDA to predict if this trader is an active trader or a passive trader. 

Assume that our dataset consists of 100 information about 100 individuals. In the dataset we have a categorical output variable $K$ and 2 numerical input variables $X$ and $Y$. Each data point contains information for $K$, $X$ and $Y$. The output variable $K$ contains 3 different types. Let’s just call them type-a, type-b and type-c. LDA will create a model system as follows:


> - When $X_a$ and $Y_a$ is from type a, $P(K_a|X_a,Y_a) \gt P(K_b|X_b,Y_b)$ and $P(K_c|X_c,Y_c)$ \\
> - When $X_b$ and $Y_b$ is from type b, $P(K_b|X_b,Y_b) \gt P(K_a|X_a,Y_a)$ and $P(K_c|X_c,Y_c)$ \\
> - When $X_c$ and $Y_c$ is from type c, $P(K_c|X_c,Y_c) \gt P(K_a|X_a,Y_a)$ and $P(K_b|X_b,Y_b)$

LDA supplies a model which creates a system to tell us the conditional probability of the data point being of each type given the input data . Furthermore, the model assigns the type with the highest conditional probability as the type from which input data comes from. From our above system, $X$ and $Y$ are assigned to type-a group when the conditional probability of being type-a is the highest compared to the conditional probability of being type-b or being type-c.

How do we get the conditional probability for each type? We will use Bayes’ theorem and Bayesian statistics. For the conditional probability $P(K_a|X_a,Y_a) $, we can rewrite it as follows:

$$ P(K_a | X_a,Y_a) = \frac{f(X_a,Y_a|K_a)P(K_a)}{\sum_{i=a,b,c} f(X_i,Y_i | K_i) P(K_i)} $$

In Bayesian statistics, 

> $P(K_a|X_a,Y_a)$ is known as the posterior probability <br>
> $f(X_a,Y_a|K_a)$ is called the likelihood function and <br>
> $P(K_a)$ is the prior probability. <br>


Since $\sum_{i=a,b,c}^{}P(K_i)$ is the same for each conditional probability, we can drop it for the ease of calculation. Hence, we need to estimate $f(X,Y|K)P(K)$ for each type. 

To estimate $P(K)$, we will calculate the ratio of number of data points of certain type over number of all data points from the dataset. The sum of $P(K)$ of all types should be $1$. 

For estimation of $f(X,Y|K)$, we need to have sample mean and variance-covariance matrix for each type from the dataset. One key point here is variance-covariance matrices from different types should all be the same for LDA. If variance-covariance matrices are different among types, then we need to use another model. We can use the dataset we get to calculate sample means and sample variance-covariance matrices for 3 types. 


**Figure 2. Scatter Plot for Input Variables X and Y and Output Variable K with 3 Types**

image.png


Figure 2 illustrates what we just discussed about. $\bar{X},\bar{Y}$ are the sample means in each type. $\sum$ is the variance-covariance matrix in each type. We require the following assumptions so we can apply LDA.

> **a.** $(X|type)$ and $(Y|type)$ are normally distributed <br>
> **b.** Variance-Covariance matrices for different types are the same <br>
> **c.** Each data point is independent of one another <br>

With the above assumptions and the values of sample means and sample variance-covariance matrices, we can estimate the likelihood functions. With estimated likelihood function and prior probability, we can obtain an estimated posterior conditional probability for each type given the input data. In LDA, however, we use the above concept and we can derive 3 Linear Discriminant functions that are related to the 3 posterior conditional probability density functions. The 3 linear discriminant functions are shown as follows:

$$ \begin{align*}
  L_a  &= \beta_{a,0} + \beta_{a,1} X + \beta_{a,2} Y  \\
  L_b  &= \beta_{b,0} + \beta_{b,1} X + \beta_{b,2} Y  \\
  L_c  &= \beta_{c,0} + \beta_{c,1} X + \beta_{c,2} Y
\end{align*} $$

Where 

$$ \begin{pmatrix}
  \beta_{a,0} & \beta_{a,1} & \beta_{a,2} \\
  \beta_{b,0} & \beta_{b,1} & \beta_{b,2} \\
  \beta_{c,0} & \beta_{c,1} & \beta_{c,2} 
\end{pmatrix} $$ 

is derived from sample means and sample variance-covariance matrices

The way to use linear discriminant functions is similar to use 3 posterior probability densities. If you have an untyped subject with data $X$ and $Y$, you plug in $X$ and $Y$ in all 3 linear discriminant functions. The linear discriminant function that gives the highest output number indicates subject is from the type of the **linear discriminant function**. With linear discriminant functions, we can also derive posterior probabilities. 

This is a very brief introduction to linear discriminant function. All the detail information about LDA will be in the required reading for this lesson. LDA gives us an idea of  how classification can be done with a supervised model. There are other classification models that we will discuss in Machine Learning module.


### **3.2. Linear Discriminant Analysis (LDA): An Application**

In this section, we provide a simple example of an application of  LDA. Let’s continue with our trader type example introduced earlier.

Assume we have the following linear discriminant functions for active traders and passive traders:

$$ L_{active} = 132 - 1.42 \times \text{Holding Time} + 1.35 \times \text{Number of Trades} $$

$$ L_{passive} = 211 + 0.52 \times \text{Holding Time} - 2.02 \times \text{Number of Trades} $$


Now we have a new trader and we would like to know this trader’s type. This trader’s holding time is 25 days and number of trades is 32 trades. Let’s plug these numbers in to the two linear discriminant functions.

$$ L_{active}= 132 – 1.42*25+1.35*32 = 139.7 $$

$$ L_{passive}= 142 +0.52*25-2.02*32 = 90.36 $$


From the above linear discriminant functions, this trader is an active trader because $L_{active} > L_{passive}$. 


## **4. Cluster Analysis**

In this section, we will talk about  a popular method in unsupervised learning modeling group: **Cluster Analysis**. Unlike classification analysis, there is no categorical output variable in the dataset for cluster analysis. Cluster analysis analyze unlabeled dataset to find similarity among data points and create clusters from the data accordingly. One example is that you have 100 stocks and for each stock we know: company name, type of company, annual sales, margin, historical price, PE ratios etc… With all this information, we would like to segment the 100 stocks into clusters of similar characteristics. We don’t have any preset view on these 100 stocks and we just want to use the available data to see if the data can offer insights on similarity in these stocks. 

There are two methods to approach cluster analysis: **Hierarchical Clustering** and **K-Means Procedure**. 


### **4.1. Hierarchical Clustering**

Hierarchical Clustering begins with treating each data point as a cluster. It then uses a **linkage function** to decide which two data points are close to each other and combine them as a new cluster. Hierarchical clustering repeats the process until all data points are in one cluster. During this merge process, we can draw the following hierarchical clustering dendrogram. 


**Figure 3. Dendrogram Example**

image.png


Figure 3 shows an example for dendrogram. We can see the process for Hierarchical clustering is to start from the bottom and merge data points that are closest to each other. It then continues to merge the new clusters with other data points or clusters until it forms a big cluster for all data points. The next step is to decide how many clusters to form for these data points.


**Figure 4. Dendrogram with Possible Cluster Choices**

image.png


Choosing how many clusters to accept is more of an art than a science. In Figure 4, we show two options for the same dendrogram. The dendrogram on the left is divided in  3 clusters and the dendrogram on the right is divided in 2 clusters. We first draw a red line across the dendrogram. On the left dendrogram, the red line cuts across 3 vertical lines. All the data points connected to this vertical line forms a cluster. 3 vertical lines form 3 clusters as shown in Figure 4. We do the same thing for the dendrogram on the right except this time we draw the red line higher. Because we draw the red line higher it only cuts across 2 vertical lines. 2 vertical lines form 2 clusters. All the data points connected to a vertical line belong to the same cluster. 

The way we check if our cluster choice makes sense is to investigate the common features in each cluster and see if we can identify distinct difference among different clusters. This is a trial-and-error process. We repeat the exercise with different numbers of clusters and then select the number that most suitable for the purpose of the project goal.


### **4.2. $K$-Means Procedure**

Another popular clustering analysis method is **$K$-Means Procedure**. $K$ in $K$-Means means the number of clusters to choose. For this method, we need to pre-select how many clusters we would like to have. Usually, we would select a few options and run through $K$-Means for each option. Then then we would conduct common features investigation for the cluster result as we describe in Hierarchical clustering. For example, for a clustering project, one could choose 3 clusters, 4 clusters and 5 clusters to run $K$-Means. One would run 3-cluster $K$-Means procedure, generate 3 clusters, investigate the features in 3 clusters to see if there are distinct differences among clusters and take a note. Then one would move to a 4-cluster $K$-Means procedure and repeat the same exercise until all cluster options are done. Then choose the number of clusters that provide the most insights for the project goal.

How do we run $K$-Means procedure? Assume we are running 4-cluster $K$-Means procedure. $K$-Means will initially randomly split data points into 4 clusters. It will then calculate the mean of each cluster. Say we have cluster $A$, cluster $B$, cluster $C$ and cluster $D$. The means for them are $\bar{A}$, $\bar{B}$, $\bar{C}$ and $\bar{D}$. Now we pick a data point $a_1$ from cluster $A$ and calculate this point’s distance to $\bar{A}$, $\bar{B}$, $\bar{C}$ and $\bar{D}$. If the distance between $a_1$ and $\bar{A}$ is actually longer than the distance between $a_1$ and $\bar{C}$, then we will reassign $a_1$ to cluster $C$. Then we recalculate the means of the 4 clusters and do the distance check again. We repeat the process until no data point being reassigned to a different cluster. This is the idea of how $K$-Means procedure runs. In the Machine Learning class, we will dive into the details of the $K$-Means algorithm.
