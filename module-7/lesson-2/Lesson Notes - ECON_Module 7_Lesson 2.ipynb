*In this lesson and the next one, we will be talking about a new and popular model to analyze complex systems: Agent-Based Modeling (ABM). One of the key differences of agent-based modeling is it models individuals or agents' behavior in the system instead of the aggregated behavior of all individuals in traditional modeling approaches. Since agent-based modeling analyzes the impact of each individual’s actions in the system, this is a computer-intensive method. In this lesson, we will talk about the general concept, the component properties, and the model structure in agent-based modeling. We then provide a very simple example to give some idea of how agent-based modeling works. This lesson and the next lesson will cover our agent-based modeling topic.*

## **1. Agent-Based Modeling Overview**

**Agent-based model and simulation (ABM or ABMS)** aims to model a complex system and obtain insights from this system. The complex system contains autonomous but diverse agents and an environment where agents can interact with each other. Each agent has their own behavioral characteristics and their behaviors will evolve from interacting with other agents and learning their behaviors. ABM uses a Monte Carlo simulation to simulate the complex system by following the rules set up for agents' attributes, the environment, ways of agents' interactions, and possibly agents' learning mechanisms. The researcher can then conduct analysis on the simulated results. 

From the above description, we can easily determine that ABM approaches modeling in a different manner than the traditional modeling methods we have learned so far. Traditional statistical methods assume there is a representative agent, say a consumer or a trader, and that this representative agent has the same characteristics or behaviors as all other agents in the model. To create a representative agent, we usually apply commonly observed characteristics or behaviors from different agents in real world to this representative agent. We then assume all agents behave like this representative agent and use this assumption for modeling. This is called a top-down approach to modeling. Regression analysis for consumer demand is a good example. To create a consumer demand regression model, we assume all the consumers have the same utility function and they all want to maximize their utility in a competitive market. We also assume all consumers are rational and always use price as the only tool to evaluate a product. The **top-down** modeling approach or macro approach apply macro-level information to a representative agent. This method allows for a simpler model for analysis purposes, but it misses the differences among agents and how their behaviors can impact other agents.

Compared to the traditional approach described in the previous paragraph, ABM approaches modeling in a very different manner. In fact, ABM starts by assuming that agents have different attributes or behaviors and that agents interact with one another inside an assigned environment and adapt to or learn from other agents' behaviors. Once the pre-designed rules are set for different agents, the environment, and interactions, the model will let the computer simulate agents' activities in the environment and show emerging phenomena. This modeling approach starting from individual agents and searching for emerging phenomena from post-model simulation is called a **bottom-up** approach. With a bottom-up approach, we are able to relax the restriction that each agent is the same. ABM explains the behavior of the system by simulating the behaviors of different agents in the system. Through repeated simulated interaction of heterogeneous agents, even with simple rules, ABM can reveal interesting outcomes that are observed in the real world. ABM focuses on individual agents' behaviors, which is more flexible and granular than a traditional modeling approach. 

Another key difference between traditional modeling and ABM is that the latter employs Monte Carlo simulation. In this way, ABM can build a more complex model for analysis. In traditional modeling, in order to build a more granular model, we may have to build several regression models and estimate them simultaneously. Many times, this system is too complicated for obtaining a closed-form analytical solution. This is where we can use a simulation to get results. With ABM, when the model structure is set up, the Monte Carlo simulation can generate several simulated outcomes for investigators to analyze. The prerequisite for running a simulation for ABM is that one must have considerable computing power to support a large number of simulations and generate outcomes. Hence, ABM requires a more powerful computational infrastructure to run. In order to create an ABM, a researcher must also have good programming skills, be able to select suitable attributes to assign to agents, and be in a position to explain the simulation outcome carefully.

Now we know the differences between a traditional regression model and an ABM. Since ABM is very different from traditional statistical models and requires coding skills to implement the model, we will use an example to explain the ABM structure with some coding to give you an idea of how ABM works. The example we will use is the segregation model proposed by Thomas Schelling. We will show implementation for ABM in the following sections. The code in the example is generic in nature, so it should be easy to convert to other programming languages. First, we will explain what the segregation model is.


## **2. Thomas Schelling’s Model of Segregation**

American economist Thomas Schelling (1997) introduced a model to analyze why people of similar backgrounds, like ethnicity, race, or socioeconomic status tend to live close to each other. The model shows that even with people who do not mind living with people from other backgrounds, over time, people still choose to move to area with people of the same background. We will use the following simplified version of the segregation model with two different groups of people as our base model to explain agent-based modeling. 

Assume we have two groups of people, group X and group Y. The people of these two groups are randomly assigned to a cell in a grid. There are also empty cells in the grid that no one is assigned to. The following figure illustrates the concept.


**Figure 1: Segregation Grid with Group X and Group Y**

image.png


In figure 1, we can see there are three types of cell status on the grid: the ones occupied by X, the ones occupied by Y and the empty ones. A person is always surrounded by any of these three types. In the segregation model, this person will get satisfaction from how many people from the same group live around them. If this person is not satisfied with the current location, this person will move. The model result states that even if the person’s satisfaction level is not high, there will still be highly segregated communities in society. 

We now have an idea of what the segregation model is doing. Let’s get into some details. First, we need to define a way to identify the neighbors. One common method to use is called Moore Neighborhood. Let’s use the following figure to explain Moore Neighborhood.


**Figure 2: Moore Neighborhood Example with $r = 0$**

image.png


From figure 2, we identify neighbors for Person A as people living adjacent to Person A. This means there are no other neighbors between Person A and their neighbors. This is the Moore Neighborhood with $r = 0$. We can extend the idea to $r = 1$. The following figure shows the case when $r=1$.


**Figure 3: Moore Neighborhood Example with $r = 1$**

image.png


We can see from Figure 3 when $r = 1$ not only is the person adjacent to Person A a neighbor but also the person that is one neighbor away from Person A is also a neighbor. We can use the same logic and identify neighbors when $r = 2$. In this lesson, we are going to use $r = 0$ as our neighbor identification rule.

Now we know how to identify neighbors. Our next step is to define a person’s satisfaction with their current location. In the model, a satisfied person is defined as surrounded by at least $s$ percent of people that are from their group. $s$ can be seen as a satisfaction benchmark. If a person has a higher percentage of people from their group as neighbors than $s$, then this person is satisfied. Otherwise, this person will move. We can see that the higher the satisfaction benchmark is, the more likely this person will be less satisfied and will move. The model assumes everyone has the same satisfaction benchmark. Let’s use the following example to explain this concept.


**Figure 4: Neighbors for Y in the Pink Cell**

image.png


In figure 4, the Y in pink cell has three neighbors. Two of the three are from the same group. Hence, this agent Y has 2/3 = 66% of neighbors from their own group. If the satisfaction benchmark were $s$ = 25%, then agent Y who is located in the pink cell will be satisfied and will not move. However, if the satisfaction benchmark were $s$ = 75% instead, then agent Y will not be satisfied and will move. 

After setting up the rules to identify neighbors and rules to relocate, we will start simulating each person’s relocation decision for several rounds until everyone is satisfied with their location. In each round, the computer program will go through each person’s neighbor satisfaction status and decide if the person needs to move. A dissatisfied person will move while a satisfied one will stay. Then, the program will move on to the next person until all people are checked. The process is then repeated in the next round until all people are satisfied and there is no need for anyone to move further.

Once everyone is satisfied with their location given a satisfaction level, we will have information to analyze the dynamics of people’s movement and the speed to get everyone satisfied with their location. From there, we will be able to get model results and draw conclusions. 


**Figure 5: Segregation Model with Satisfaction Benchmark = 0.6**

image.png



Figure 5 shows the grid of the initial stage of the simulation and the grid of the final stage of the simulation for a satisfaction benchmark of 0.6. The red and blue cells on the grid represent group X and Y respectively, and the white cells are empty cells. We can see in the finishing stage of the grid that there are red and blue clusters around the grid. It means there is a certain degree of segregation formed autonomously from group X and Y. This is one way to present the simulation result. We will dive deeper to show more ways to present the result with this example in the next lesson.

## **3. Agent-Based Model Components and Structure**

A typical agent-based model will have the following three key components:

> a. A group of agents and their behaviors and attributes <br>
> b. The environment in which agents interact with one another <br>
> c. A mechanism that outlines an agent's position and how agents interact with one another. This mechanism can also include how agents adapt in learning others’ behaviors. <br>


**Figure 6: Plot of Typical Components in an Agent-Based Model**

image.png


In figure 5, we show a scheme of a typical agent-based model. A researcher has to identify these three components, set up the model and write a computer program for interactions among agents and between agents and the environment. Once the above steps are ready, the researcher will simulate the model using a Monte Carlo simulation repeatedly to obtain a sufficient amount of simulated results. The researcher will then analyze the results and look for any emerging insights. The simulation process usually works through a timeline assigned before running the simulation. However, this is not necessarily the only case: The simulation can also run according to an activity or event-based schedule. Since ABM is a simulation-based model, this modeling structure does not provide a forecasting function as traditional models do. The focus of agent-based modeling is more on discovering emergent insights from simulated outcomes than on prediction.
